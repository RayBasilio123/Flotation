```python
import pandas as pd

# Caminho do arquivo
caminho_arquivo = r"c:\Users\rayba\Music\mestrado_dados\vale-nova\dados\Flotação_Dados_Primeiro_Semestre.xlsx"

# 1) Ler o arquivo, pulando as 5 primeiras linhas
df = pd.read_excel(caminho_arquivo, skiprows=5)

# 2) Remover as colunas completamente vazias
df.dropna(axis=1, how="all", inplace=True)

# 3) Dicionário de renomeação revisado/conforme as colunas que aparecem
rename_dict = {
    # Se desejar renomear "Inicio" e "Fim" para algo mais conciso
    "Inicio": "inicio",
    "Fim": "fim",
    
    # Adicione as colunas que realmente aparecem:
    "P_CONFLTTO_QQ_GLOBAL_SIO2": "conc_silica_alimentacao",
    "P_CONFLTTO_QQ_GLOBAL_FE":   "conc_fe_alimentacao",
    
    # Já renomeadas anteriormente
    "TO_PHY4804_02":    "param_dosagem_amido",
    "TO_PHY8271_01":    "dosagem_amina_conc_magnetica",
    "TO_AI6401_01":     "ph_flotacao_linha01",
    "TO_AI6402_01":     "ph_flotacao_linha02",
    "TO_DI6311_01":     "densidade_alimentacao_flotacao",
    
    "TO_LI6401_01":     "nivel_celula_li640101",
    "TO_LI6402_01":     "nivel_celula_li640201",
    "TO_LI6403_01":     "nivel_celula_li640301",
    "TO_LI6402_02":     "nivel_celula_li640202",  # se for nível da célula 640202
    "TO_LI6411_01":     "nivel_celula_li641101",
    "TO_LI6411_02":     "nivel_celula_li641102",
    "TO_LI6412_01":     "nivel_celula_li641201",
    "TO_LI6412_02":     "nivel_celula_li641202",
    "TO_LI6421_01":     "nivel_celula_li642101",
    "TO_LI6422_01":     "nivel_celula_li642201",
    "TO_LI6431_01":     "nivel_celula_li643101",
    "TO_LI6432_01":     "nivel_celula_li643201",
    "TO_FI6311_01":     "vazao_alimentacao_flotacao",
    
    # Se você tiver tags de imagem:
    "TO_MF641102_IMG_RED":   "cor_641102_red",
    "TO_MF641102_IMG_GREEN": "cor_641102_green",
    "TO_MF641102_IMG_BLUE":  "cor_641102_blue",
    "TO_MF641202_IMG_RED":   "cor_641202_red",
    "TO_MF641202_IMG_GREEN": "cor_641202_green",
    "TO_MF641202_IMG_BLUE":  "cor_641202_blue",
}

# 4) Aplicar a renomeação
df.rename(columns=rename_dict, inplace=True)

# 5) Conferir resultado
print("Colunas após renomeação:")
print(df.columns.tolist())

print("\nInício do DataFrame:")
print(df.head())
```
